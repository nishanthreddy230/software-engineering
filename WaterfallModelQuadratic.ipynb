```python
import numpy as np
import matplotlib.pyplot as plt

# Quadratic function
def quadratic(x, a=-0.8, b=0, c=10):
    return a * x**2 + b * x + c

# Waterfall stages
stages = ["Requirements", "Design", "Implementation", "Testing", "Deployment", "Maintenance"]
x_positions = np.linspace(-2, 2, len(stages))
y_positions = quadratic(x_positions)

# Curve data
x_curve = np.linspace(-2.5, 2.5, 500)
y_curve = quadratic(x_curve)

# Create the plot
plt.figure(figsize=(12, 7))
plt.plot(x_curve, y_curve, color="blue", linestyle="--", label="Quadratic Curve")

# Plot stage points and vertical/horizontal connectors
for i in range(len(stages)):
    if i > 0:
        # Horizontal line
        plt.plot(
            [x_positions[i-1], x_positions[i]],
            [y_positions[i-1], y_positions[i-1]],
            color="blue",
            linewidth=2
        )
        # Vertical line
        plt.plot(
            [x_positions[i], x_positions[i]],
            [y_positions[i-1], y_positions[i]],
            color="blue",
            linewidth=2
        )
    # Stage markers
    plt.scatter(x_positions[i], y_positions[i], color="red", zorder=5, s=100, edgecolor="black")
    plt.text(x_positions[i], y_positions[i] + 0.4, stages[i], ha="center", fontsize=12, fontweight="bold")

# Final adjustments to the plot
plt.title("Waterfall Model Representation using Quadratic Curve", fontsize=16, fontweight="bold")
plt.xlabel("Stages (X-axis)", fontsize=12)
plt.ylabel("Value (Y-axis)", fontsize=12)
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)
plt.legend()
plt.xlim(-3, 3)
plt.ylim(min(y_curve) - 1, max(y_curve) + 2)
plt.tight_layout()

# Show the plot
plt.show()
```
